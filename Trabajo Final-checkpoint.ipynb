## Descripción del data set
El data set contiene información sobre distintos parámetros que se obtienen al realizar un estudio llamado tococardiograma fetal (CTG) en embarazadas durante su trabajo de parto. Este estudio es de bajo costo y permite detectar posibles riesgos en el feto en el momento del trabajo de parto, por lo tanto poder interpretar los resutados de los mismos colabora en la toma de decisiones a la hora del parto.

Se describen 21 atributos utilizados en la medición de la frecuencia cardíaca fetal (FHR) y las contracciones uterinas (UC) en los CTG. Los datos provienen de 2126 mujeres embarazadas que estaban en el tercer trimestre de embarazo de Portugal.
De acuerdo a estándares establecidos, las variables de riesgo utilizadas para analizar el estado del feto incluyen descripciones cuantitativas de FHR y UC. 

Las CTG de las gestantes fueron clasificadas por tres expertos especialistas en obstetricia, estableciendo cuáles eran consideradas como **1- normales, 2- sospechosos y 3- patológicos**.

Las variables del data set son las siguientes:

**Baseline value**: frecuancia cardíaca fetal en latidos por minutos.

**Accelerations**: número de aceleraciones por segundo.

**Fetal_movement**: número de movimientos fetales por segundo.

**Uterine_contractions**: número de contracciones uterinas por segundo.

**Light_decelerations**: número de desaceleraciones de luz por segundo.

**Severe_decelerations**: número de desaceleraciones severas por segundo.

**Prolongued_decelerations**: número de desaceleraciones prolongadas por segundo.

**Abnormal_short_term_varability**:porcentaje de tiempo con variabilidad anormal.

**Mean_value_of_short_term_variability**: Promedio de variabilidad de corto plazo.

**Percentage_of_time_with_abnormal_long_term_variability**:Porcentaje de tiempo con variabilidad anormal de largo plazo.

**Mean_value_of_long_term_variability**:Valor promedio de variabilidad a largo plazo.

**Histogram_width**: Ancho del histograma de FHR.

**Histogram_min**:Mínimo del histograma de FHR.

**Histogram_max**: Máximo del histograma de FHR.

**Histogram_number_of_peaks**: Número de picos del histograma.

**Histogram_number_of_zeroes**: Número de ceros del histograma.

**Histogram_mode**: Moda del histograma.

**Histogram_mean**:Media del histograma.

**Histogram_median**: Mediana del histograma.

**Histogram_variance**: Variancia del histograma.

**Histogram_tendency**: Tendencia del histograma.

**Fetal_health**: Clase según el estado fetal.


El **objetivo** de este trabajo es emplear el el data set para desarrollar un modelo de aprendizaje automático que pueda clasificar fetos de alto riesgo (tanto sospechosos como patológicos) con tanta precisión como profesionales médicos altamente capacitados.

A continuación, con el objetivo de unificar aquellas variables muy correlacionadas, para ver si se podía, a nivel interpretativo simplificar su aporte al data set, se realizó un **Análisis de Componentes Principales (PCA)**, teniendo en cuenta que en realidad el número de variables no es tan elevada como para precisar una técnica de reducción de dimensionalidad. De todas maneras, a priori se puede ver que el resto de las variables guardan poca correlación entre ellas, y por ende es muy difícil que un PCA logre efectivamente reducir el número de variables necesarias.


```python
#Se realiza un PCA
# Estandarizamos las variables
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_std = sc.fit_transform(ds_sinclase)
print(np.mean(x_std))
print(np.std(x_std)) #El prom debe ser 0 y el sd 1
```

```python
# realizamos un analisis de componentes principales para 4 componentes 
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
pca.fit_transform(x_std)

# nos basamos en el scree plot para ver con cuantas variables nos quedamos
percent_var = np.round(pca.explained_variance_ratio_*100, decimals=1)   #calculamos el porcentaje de la variación para cada componente
labels = ['PC'+str(p) for p in range(1,len(percent_var)+1)]

# grafico para ver la contibución de cada uno de los componentes
f, ax = plt.subplots(figsize=(8,5))
plt.bar(x=range(1,len(percent_var)+1), height=percent_var, tick_label=labels)
plt.xlabel('Componente Principal')
plt.ylabel('Porcentage de la Variación que explica')
plt.title('Scree Plot mostrando los 4 primeros componentes')
plt.savefig("PCA.jpg", bbox_inches='tight')
plt.show()
```

```python
explained_var = pca.explained_variance_ratio_
explained_var
```

```python
cum_var = np.cumsum(np.round(explained_var, decimals = 3))
cum_var_perc = cum_var*100
```

```python
# creamos un dataframe con los 3 componentes
pc_df = pd.DataFrame(['PC1','PC2', 'PC3', "PC4"], columns=['PC'])
explained_var_df = pd.DataFrame(explained_var, columns=['Explained variance'])
cum_var_df = pd.DataFrame(cum_var_perc, columns=['Cumulative variance (in %)'])
```

```python
total_var_explained = pd.concat([pc_df, explained_var_df, cum_var_df], axis=1)
total_var_explained
```

Lo que puedo observar es que, tal como se esperaba, la variancia acumulada con la componente número 4 es del 61.4%, esto lo que indica es que para poder ver representado un porcentaje relevante de variabilidad de datos, que justifique la aplicación del PCA debo emplear un número de componentes principales tales que hacen que no tenga sentido aplicar este método de reducción de la dimensionalidad con este data set. Se concluye no proseguir por este camino.

## Tratamiento del desbalanceo de las clases
Para solucionar el desbalanceo de las clases se optó por probar tres métodos. 

1- En primera instancia con **Oversampling de la clase minoritaria**. Se empleará el método **RandomOversample** de la librería **imblearn**.

2- Luego se intentará realizar el balanceo aplicando **Undersampling** y disminuyendo así la clase mayoritaria.

3- Se aplicará en simultáneo un algoritmo de undersampling y otro de oversampling a la vez al dataset. **SMOTE**(Synthetic Minority Oversampling TEchnique) para oversampling y **Tomek** para undersampling.
